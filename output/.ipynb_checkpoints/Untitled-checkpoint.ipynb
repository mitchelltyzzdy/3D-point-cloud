```python
def check_duplicate(output_df):
    """Because the way I generate ouput file in c++, there may be duplicated points for decompressed data points(column3,4,5).
    Check if any two row has the same decompressed data points
    """
    
    A = output_df.values  # Convert to 2D array first because operating in dataframe is much slower.
    
    count = 0
    for i in range(len(output_df)):
        for j in range(i+1, len(output_df)):
            if ((A[i][3]==A[j][3]) & (A[i][4]==A[j][4]) & (A[i][5]==A[j][5])):
                #print("i is:" +str(i) )
                #print("j is:" +str(j) )
                count = count+1
                
    if count == 0:
        print("There's no duplicated points")
    else:
        print("there are "+ str(count) + "duplicated points")
    
```

```python
check_duplicate(df_A)
```

```python
check_duplicate(df_B)
```

```python
check_duplicate(df_C)
```
